```python
import numpy as np
import matplotlib.pyplot as plt

# Function 1
def f1(x1, x2):
    return 100 * x2**2 + 100 * x1**4 - 200 * x2 * x1**2 + x1**2 - 2 * x1 + 1

# Function 2
def f2(x1, x2):
    return x1**4 + x2**4 - 20 * x1**3 - 20 * x2**3 + 100 * x1**2 + 100 * x2**2

# Function 3
def f3(x1, x2):
    return -np.cos(x1) * np.cos(x2) * np.exp(-((x1 - np.pi)**2 + (x2 - np.pi)**2))

# Function 4
def f4(x1, x2):
    return 2 * x1**2 - 1.05 * x1**4 + x1**6/6 + x1 * x2 + x2**2

# Create a grid of values
x1 = np.linspace(-4, 4, 400)
x2 = np.linspace(-4, 4, 400)
X1, X2 = np.meshgrid(x1, x2)

# Plotting level sets for each function
plt.figure(figsize=(15, 10))

# Function 1
plt.subplot(221)
plt.contour(X1, X2, f1(X1, X2), levels=np.logspace(-1, 4, 20))
plt.title('Level Sets for f1(x)')

# Function 2
plt.subplot(222)
plt.contour(X1, X2, f2(X1, X2), levels=np.logspace(-1, 4, 20))
plt.title('Level Sets for f2(x)')

# Function 3
plt.subplot(223)
plt.contour(X1, X2, f3(X1, X2), levels=np.logspace(-4, 4, 20))
plt.title('Level Sets for f3(x)')

# Function 4
plt.subplot(224)
plt.contour(X1, X2, f4(X1, X2), levels=np.logspace(-1, 4, 20))
plt.title('Level Sets for f4(x)')

plt.show()
```

#Observations:

#1: Level sets for quadratic functions (like f1 and f2) show elliptical shapes. They are useful for identifying the minimum, which is the center of the ellipse.
#2 Level sets for oscillatory functions (like f3) can be complex, and the minima are often located in regions where the oscillations are minimal.
#3 Level sets for a combination of quadratic and non-quadratic terms (like f4) may exhibit intricate patterns. Identifying the minimum involves considering the interplay of different terms.
##To analyze the minima, you can use first and second-order optimality conditions such as gradient descent and Hessian matrix analysis. These conditions help in locating critical points and determining whether they are minima, maxima, or saddle points. The shape of the level sets provides a visual aid in understanding the landscape of the optimization problem.
