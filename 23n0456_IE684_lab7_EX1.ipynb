```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss functions
def Gh(z):
    return np.maximum(0, 1 - z)

def Gl(z):
    return np.log(1 + np.exp(-z))

def Gsh(z):
    return np.maximum(0, 1 - z)**2

# Generate z values
z_values = np.linspace(-5, 5, 10000)

# Compute the loss for each function
loss_hinge = Gh(z_values)
loss_logistic = Gl(z_values)
loss_squared_hinge = Gsh(z_values)

# Plot the loss functions
plt.figure(figsize=(10, 6))
plt.plot(z_values, loss_hinge, label='Hinge Loss', color='blue')
plt.plot(z_values, loss_logistic, label='Logistic Loss', color='red')
plt.plot(z_values, loss_squared_hinge, label='Squared Hinge Loss', color='green')

plt.title('Loss Functions')
plt.xlabel('z')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

**Hinge Loss (Lh):**

Penalizes misclassifications linearly with the margin z, focusing on maximizing the margin to the correct class while ignoring classification when the margin exceeds 1.

**Logistic Loss (Ll):**

Smooth, convex function penalizing misclassifications using the logistic function, providing diminishing returns as z moves away from zero.

**Squared Hinge Loss (Lsh):**

Quadratically penalizes misclassifications, causing a more severe penalty for large margin violations compared to hinge loss, with a continuous but non-differentiable behavior at z = 1.
